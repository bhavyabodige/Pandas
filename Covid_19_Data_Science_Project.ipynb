Form a subset of data df_covid19_country using Group by



```python
# The groupby command allows us to divide our data into different groups and perform our data analysis.

df_covid19_country  = df_covid19.groupby(['continent','location','date'],as_index = False)[['total_cases','total_deaths','total_vaccinations']].sum()
```

```python
# Dataframe which gives the total cases, deaths and vaccinations group by continent, location and date
df_covid19_country
```

We will find the total countries where total_cases is greater than 200000.



```python
df1 = df_covid19_country[df_covid19_country['total_cases'] > 200000]
# Number of countries where total cases > 200000
#n = df1['location'].unique()
#len(n)
     # or 
df1['location'].nunique()
```

```python
# Names of countries where total cases > 200000
countries = df1['location'].unique()
print("Countries where total cases are more than 200000 are :",countries)
```

We will look at how many dates we have in total where total_cases is greater than 200000

```python
# # Different dates where total_cases is greater than 200000
dates_cases = df1['date'].unique()
print("The dates where total cases are more than 200000 are :",dates_cases)
```

Finding the continent with highest positive rate


```python
df_covid19.update(df_covid19['positive_rate'].fillna(0)) # Replacing the null values by 0
dfp = df_covid19[df_covid19['positive_rate'] > 0]
# Finding the highest positivity rate 
dfp['positive_rate'].max()
```

```python
# Finding the continent with highest positivity rate
df3 = dfp.loc[dfp['positive_rate'] == 0.9666, 'continent']
df3
```

Drawing a bar plot using continents and their positivity rate



```python
## Plotting a graph of the continents and their positive rates

sns.barplot(data=dfp, x='continent', y='positive_rate') ## uses the dataframe with appropriate x and y values
plt.title("Continents and their positivity rate")  # Produces a title for the graph
plt.ylabel("Positivity rate") # Create a y axis label
plt.xlabel("Continent") # Create an x axis label
plt.show() ## show the plot
```

Finding the continents, countries and total number of deaths where diabetes prevalence is more than 5

```python
df5 = df_covid19[df_covid19['diabetes_prevalence']> 5]
# Continents and Countries with diabetes prevalence more than 5
diabetes_continents = df5['continent'].unique()
print("The continents where the diabetes prevalence is more than 5 are :",diabetes_continents)
diabetes_countries = df5['location'].unique()
print("The countries where the diabetes prevalence is more than 5 are :",diabetes_countries)
```

```python
#  Create a data frame that groups the columns continent, location and diabetes prevalence with more than 5 against the Total deaths
diabetes_data = df5.groupby(["continent","location","diabetes_prevalence"],as_index=False)["total_deaths"].sum()
# Finding the continents, countries and total number of deaths where diabetes prevalence is more than 5
diabetes_data
```

Bar plot showing the continents 
where diabetes prevalance is more than 5

```python
## Plotting a graph of the continents and their total deaths 

sns.barplot(data=diabetes_data, x='continent', y='diabetes_prevalence') ## uses the dataframe with appropriate x and y values
plt.title("Continents where diabetes prevalance is more than 5")  # Produces a title for the graph
plt.ylabel("Diabetes Prevalence") # Create a y axis label
plt.xlabel("Continent") # Create an x axis label
plt.show() ## show the plot
```
